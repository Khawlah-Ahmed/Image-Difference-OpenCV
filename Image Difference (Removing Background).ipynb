```python
"""
Khawla Alarfaj | 215027377
Image Difference

"""
import cv2
import numpy as np
from matplotlib import pyplot as plt
orgImage = cv2.imread("without man.png")
orgObjImage = cv2.imread("with man.png")
result = cv2.absdiff(orgObjImage, orgImage)
image = cv2.cvtColor(result, cv2.COLOR_BGR2BGRA) # to have 4 channels (Blue, Green, Red, and Alpha) 
for i in range(0, 25, 1): # Note: 3 level nested loop would take time (about 2 minutes)
    for j in range(0, 25, 1):
        for k in range(0, 25, 1):
            bg_color = np.array([i,j,k, 255])
            mask = np.all(image == bg_color, axis=2)
            image[mask] = [0, 0, 0, 0]          
cv2.imwrite("Transparent Background.png", image)
image = cv2.cvtColor(image, cv2.COLOR_BGRA2RGBA) # since plt.show() shows the image as RGBA
plt.imshow(image)
cv2.destroyAllWindows()
```
